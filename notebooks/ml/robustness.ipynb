## Testing robustness of models

```python
version='v2.11'
norm='minmax'

dump='../../dump_seeds'

import sys,os
sys.path.append(os.path.abspath('../../src/ml'))
import pandas as pd
```

```python
results_path='../../results/score_tables/'

avg_df=pd.read_csv(os.path.join(results_path,f'scores_avg_seeds.csv'))
std_df=pd.read_csv(os.path.join(results_path,f'scores_std_seeds.csv'))
all_df=pd.read_csv(os.path.join(results_path,f'scores_all_seeds.csv'))
```

```python
avg_df = avg_df[
    (avg_df['version'] == version) &
    (avg_df['normalization'] == norm) &
    (avg_df['model'] != 'pytorch_mlp') &
    (avg_df['input'].str.contains('protein'))
]
all_df = all_df[
    (all_df['version'] == version) &
    (all_df['normalization'] == norm) &
    (all_df['model'] != 'pytorch_mlp') &
    (all_df['input'].str.contains('protein'))
]
std_df = std_df[
    (std_df['version'] == version) &
    (std_df['normalization'] == norm) &
    (std_df['model'] != 'pytorch_mlp') &
    (std_df['input'].str.contains('protein'))
]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

metric = 'balanced_accuracy'

# Combine model and input for x-axis labels
all_df['model_input'] = all_df['model'] + ' - ' + all_df['input']

sns.set(style="whitegrid")  # clean background

# Color palette based on dataset (input)
datasets = all_df['input'].unique()
palette = dict(zip(datasets, sns.color_palette("tab10", n_colors=len(datasets))))

plt.figure(figsize=(14,6))

# Boxplot with color based on dataset
ax = sns.boxplot(
    data=all_df,
    x='model_input',
    y=metric,
    hue='input',  # color by dataset
    palette=palette,
    showfliers=False
)

# Overlay swarmplot for individual points, color by dataset
sns.swarmplot(
    data=all_df,
    x='model_input',
    y=metric,
    hue='input',
    palette=palette,
    dodge=True,  # match hue separation
    size=4,
    alpha=0.7
)

# Title and labels
plt.title(f'{metric} Across Models and Input Datasets\n(version={version}, normalization={norm})', fontsize=14, weight='bold')
plt.xlabel('Model - Input Dataset', fontsize=12)
plt.ylabel(metric, fontsize=12)

# Rotate x-axis labels
plt.xticks(rotation=45, ha='right', fontsize=10)

# Remove duplicate legend from swarmplot
handles, labels = ax.get_legend_handles_labels()
plt.legend(handles[:len(datasets)], labels[:len(datasets)], title='Dataset', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
